## how to use getWholeGenomeVariantsInRegion

This function takes 
  - an implicit argument, the genomic region currently displayed in igv (above): *roi* - the region of interest
  - *altToRefRato*: a minimum proportion of AD/control samples with het or homozygous variants at each location
  - *minAltCount*: the minimum number of AD (affected) samples at each location
  - *display*: render the variants as a track in igv
  - *color*: of the igv track
  - *trackHeight*: 50 pixels by default

## "model threatening" variants are calculated according to four independently varying parameters

1. **chromosomal region**: inferred from the region currently displayed in igv
2. **source of binding sites**: "footprints" *or* "enhancer.motifs *or* "dhs.motifs" *or* "allDNA.motifs"
3. **source of variants**: "eqtl.snps" *or* "wg.AD.variants"
4. **motif-TF-mapping table**: "MotifDb" *or* "TFClass"
5. **shoulder**: 0 or more bases 
6. **tfs**:  transcription factors to evaluate (gene symbols, usually drawn from the "gene" column of a trena model's table)
